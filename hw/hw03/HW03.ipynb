```python
# First temperature record of each region.

# Variable setting
region_time = dict()
region_temp = dict()

# Open the file.
Filename = r'.\data\64_72hr_EN.xml'
with open(Filename,'rt',encoding='utf-8') as file:
    XmlReader = dict(xmltodict.parse(file.read()))
    
# Get the list of locations. And traverse the list to get temperature of each region.
Locations = XmlReader['cwbopendata']['dataset']['locations']['location']
for local in Locations:
    temperature = local['weatherElement'][0]['time'][0]['elementValue']['value']
    deg = local['weatherElement'][0]['time'][0]['elementValue']['measures']
    time = local['weatherElement'][0]['time'][0]['dataTime']
    region_temp[local['locationName']] = temperature+' '+deg
    region_time[local['locationName']] = time


print('First temperature in each region: \n',region_temp,'\n')
print('Datatime in each region: \n',region_time)
```

```python
# All temperature record for the first region.


# Variable setting
region_time = list()
region_temp = list()


# Open the file.
Filename = r'.\data\64_72hr_EN.xml'
with open(Filename,'rt',encoding='utf-8') as file:
    XmlReader = dict(xmltodict.parse(file.read()))
    
# Get the list of first location and traverse the data in time series.
Locations = XmlReader['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][0]['time']
for t in Locations:
    time = t['dataTime']
    temp = t['elementValue']['value']
    deg = t['elementValue']['measures']
    region_time.append(time)
    region_temp.append(temp+' '+deg)


# Print the result.
print('Temperature: \n',region_temp,'\n')
print('Datatime: \n',region_time)
```
