```python
# Dcard
import requests, json, re
from bs4 import BeautifulSoup as bs
from pprint import pprint

url = 'https://www.dcard.tw/f'
header = {'User-Agent':'Mozilla/5.0'}

rs = requests.Session()
r = rs.get(url, headers = header)

# Review what I have learned (Auto-detection of the format of response content)
response_type = r.headers['content-type']
if 'html' in response_type: 
    soup = bs(r.text, 'html.parser')
    data = soup.prettify()
    print('Data type is .html\n')
    
    # Get all the links inside the .html file.
    # For dcard, the website can be accessed easily by join url[0: len(url)-2] and link[n], where n belongs to range(0,len(links)).
    links = set()
    for link in soup.find_all('a'):
        if link.get('href') not in links:
            links.add(link.get('href'))
    print('The link of each posts are: \n',links)
if 'json' in response_type:
    data = json.loads(r.text)
    print('Data type is .json\n')
```

```python
# zhihu
import requests, json, re
from bs4 import BeautifulSoup as bs
from pprint import pprint


url = 'https://www.zhihu.com/explore'
header = {'User-Agent':'Mozilla/5.0'}

rs = requests.Session()
r = rs.get(url, headers = header)
response_header = r.headers['content-type']
if 'html' in response_header:
    print('Data type is .html\n')
    soup = bs(r.text, 'html.parser')
    data = soup.prettify()
    
    # get the links
    # links in zhihu becomes more complex than dcard. 
    # For first 8 links, user needs to join 'https:' with links[n], where n belongs to range(0, 8)
    # For links in middle, user needs to join 'https:' with links[0] and links[n] , where n belongs to range(8, len(links)-21)
    # For the rest, user is able to access the website by request url as links[n], where n belongs to range(len(links)-21, len(links))
    links = set()
    for link in soup.find_all('a'):
        if link.get('href') not in links:
            links.add(link.get('href'))
    pprint(links)
if 'json' in response_header:
    print('Data type is .json\n')
    data = json.loads(r.text)
```
