```python
import requests, json

url = 'https://api.github.com/events'#'https://www.dcard.tw/f/sex'
rs = requests.Session()
r = rs.get(url)
#print("Return from request.get: \n",r.text,'\n')

j = json.loads(r.text)
#print('After json load: \n',j,'\n')

## The different between json.loads and r.text is that:
## r.text returns a str object includes all characters in the body of HTMLresponse.
## json.loads decode the .json file and return the data with the correspond-structure in python (compared to how the data saved).
print(len(r.text), len(j))
print(type(r.text), type(j))
```
